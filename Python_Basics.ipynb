```python
def factorial():  #Factorial calculation
  n = int(input("Enter the number: "))
  if(n==0):
    fact = 1
    print(fact)
  if(n<0):
    print("Factorial not defined.")
  if(n>0):
    fact = 1
    for i in range (1,n+1):
      fact = fact * i
    return fact
factorial()
```

```python
#Practice Assignment 2a: Hill Cipher Algorithm (Encryption)
import numpy as np
word = []
x = str(input("Enter the string to be encrypted into Hill Cipher:\n"))
x1 = x.upper() #Makes all characters in the string, uppercase letters
x1 = "".join(x1.split()) #Removes blank spaces, if any
for i in x1:
  k = ord(i) - 65
  word.append(k)
#print(n)
#print("\n")
#print(word)
#print("\n")
while(len(word)%3!=0): #Adding 0s so that there are 3*n elements (n is a positive integer)
  word.append(0)
n = len(word)//3 #'//' is the int division operator
word1 = np.array(word).reshape(3,n) #Forming a 3*n matrix
#print(word1)
#print("\n")
key = np.array([[6,24,1],[13,16,10],[20,17,15]]) #[[GYB],[NQK],[URP]] chosen as the key matrix
#print(key)
#print("\n")
def encrypt():
  #print("Encryted Matrix: ")
  ans = np.dot(key,word1)
  #print(ans)
  #print("\n")
  #print("Reducing the Matrix: ")
  ans1 = (ans%26) #Taking mod 26
  #print(ans1)
  #print("\n")
  #print("Converted to a List: ")
  ans2 = ans1.tolist()
  #print(ans2)
  #print("\n")
  ans3 = []
  for l in ans2[0]:
    l = l + 65
    m = chr(l)
    ans3.append(m)
  for d in ans2[1]:
    d = d + 65
    n = chr(d) 
    ans3.append(n)
  for e in ans2[2]:
    e = e + 65
    o = chr(e)
    ans3.append(o)
  #print(ans3)
  ans_f = "" #Initialised empty string
  print("Encrypted String:")
  ans_f = ans_f.join(ans3)
  print(ans_f)
encrypt()
```

```python
#Practice Assignment 2b: Hill Cipher Algorithm (Decryption)
import numpy as np
word = []
x = str(input("Enter the string to be decrypted from Hill Cipher:\n"))
x1 = x.upper() #Makes all characters in the string, upper case letters
x1 = x1.replace(" ","") #Removes blanks spaces, if any
for i in x1:
  k = ord(i) - 65
  word.append(k)
#print(n)
#print("\n")
#print(word)
#print("\n")
while(len(word)%3!=0): #Adding 0s so that there are 3*n elements (n is a positive integer)
  word.append(0)
n = len(word)//3 #'//' is the int division operator
word1 = np.array(word).reshape(3,n) #Forming a 3*n matrix
#print(word1)
#print("\n")
key = np.array([[8,5,10],[21,8,21],[21,12,8]]) #[[IFK],[VIV],[VMI]] chosen as the key matrix
#print(key)
#print("\n")
def decrypt():
  #print("Decryted Matrix: ")
  ans = np.dot(key,word1)
  #print(ans)
  #print("\n")
  #print("Reducing the Matrix: ")
  ans1 = (ans%26) #Taking mod 26
  #print(ans1)
  #print("\n")
  #print("Converted to a List: ")
  ans2 = ans1.tolist()
  #print(ans2)
  #print("\n")
  ans3 = []
  for l in ans2[0]:
    l = l + 65
    m = chr(l)
    ans3.append(m)
  for d in ans2[1]:
    d = d + 65
    n = chr(d) 
    ans3.append(n)
  for e in ans2[2]:
    e = e + 65
    o = chr(e)
    ans3.append(o)
  #print(ans3)
  ans_f = "" #Initialised empty string
  print("Decrypted string:")
  ans_f = ans_f.join(ans3)
  print(ans_f)
decrypt()
```

```python
#Practice Assignment 3: Caesar Cipher Algorithm (Encryption and Decryption)
#Caesar Cipher Encryption
def caesar_cipher_encrypt():
    
    word = input("Enter the word to be encrypted into caesar cipher: ")
    word = word.replace(" ","") #Removes white spaces, if any
    word = word.upper() #Converts all lowercase letters to uppercase, if any

    word1 = list(word.strip(" ")) #Converts string to list
    
    shift = int(input("Enter the value of the shift: "))
    
    word2 = []
    for i in word1:
        x =  (ord(i) - 65)%26 + shift  #Takes ASCII values to form the new values
        word2.append(x)
    
    encrypt = []
    for j in word2:
        j = j + 65
        y = chr(j) #Converts the new values to the corresponding alphabets
        encrypt.append(y)
    
    encrypted_word = ''.join(encrypt) #Converts the list to string of the encrypted word
    print("Encrpted word in Caesar Cipher is:" ,encrypted_word)

#Caesar Cipher Decryption
def caesar_cipher_decrypt():
    
    word = input("Enter the word to be decrypted from caesar cipher: ")
    word = word.replace(" ","") #Removes white spaces, if any
    word = word.upper() #Converts all lowercase letters to uppercase, if any

    word1 = list(word.strip(" ")) #Converts string to list
    
    shift = int(input("Enter the value of the shift: "))
    
    word2 = []
    for i in word1:
        x =  (ord(i) + 65) - shift  #Takes ASCII values to form the new values
        word2.append(x)
    
    decrypt = []
    for j in word2:
        j = j - 65
        y = chr(j) #Converts the new values to the corresponding alphabets
        decrypt.append(y)
    
    decrypted_word = ''.join(decrypt) #Converts the list to string of the encrypted word
    print("Decrpted word from Caesar Cipher is:" ,decrypted_word)

caesar_cipher_encrypt()
caesar_cipher_decrypt()
```

```python
#Sorting Technique: Bubble Sort 
arr = []
n = int(input("Enter number of elements: "))
for i in range (0,n):
  k = float(input("Enter the element: "))
  arr.append(k)
print(arr)
for j in range (0,len(arr)):
  for l in range (0,len(arr)-j-1):
    if(arr[l]>arr[l+1]):
      temp = arr[l]
      arr[l] = arr[l+1]
      arr[l+1] = temp
print("Array sorted in ascending using bubble sort:\n",arr)
#Note: For descending order: change the sign '>' to '<'.
```

```python
#Searching Technique: Linear search
arr = []
n = int(input("Enter number of elements: "))
for i in range (0,n):
  k = float(input("Enter the element: "))
  arr.append(k)
key = float(input("Enter the element to be searched: "))
for k in range(0,n):
  if (arr[k]==key):
    print(key,"found in the array.")
```
